# FINDING THE OUTLIERS WITH THE HELP OF BOXPLOT IN FINAL DATASET.

```python
num_cols=final_data.select_dtypes(include=["float64","int64"]).columns

plt.figure(figsize=(15,25))
for i,col in enumerate(num_cols,1):
    plt.subplot(5,3,i)
    sns.boxplot(y=final_data[col])
    plt.title(col)
    
plt.tight_layout()
plt.show()

#The below boxplots shows that there are outliers in "temp","atemp","casual","registered","cnt" columns.
```

# HANDELING THE OUTLIERS.

```python
# Handeling outlies for "temp" column.
q1=final_data["temp"].quantile(0.25)
q2=final_data["temp"].quantile(0.50)
q3=final_data["temp"].quantile(0.75)
print("The first quantile is",q1)
print("The second quantile is",q2)
print("The third quantile is",q3)

IQR=q3-q1
print("the Interquartile range is",IQR)

lower_limit=q1-1.5*IQR
upper_limit=q3+1.5*IQR
print("The lower limit is",lower_limit)
print("The upper limit is",upper_limit)
```

```python
final_data["temp"]=np.where(final_data["temp"]>0.36,final_data["temp"].mean(),final_data["temp"])
final_data["temp"]=np.where(final_data["temp"]< 0.04000000000000002,final_data["temp"].mean(),final_data["temp"])
```

```python
sns.boxplot(final_data["temp"])
plt.show()
```

```python
# Handeling outlies for "atemp" column.
q1=final_data["atemp"].quantile(0.25)
q2=final_data["atemp"].quantile(0.50)
q3=final_data["atemp"].quantile(0.75)
print("The first quantile is",q1)
print("The second quantile is",q2)
print("The third quantile is",q3)

IQR=q3-q1
print("the Interquartile range is",IQR)

lower_limit=q1-1.5*IQR
upper_limit=q3+1.5*IQR
print("The lower limit is",lower_limit)
print("The upper limit is",upper_limit)
```
